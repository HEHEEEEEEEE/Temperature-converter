```python
from tkinter import *
from tkinter.messagebox import *


root = Tk()
root.title("TIC-TAC-TOE")


digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]


mark = ''


count = 0


panels = ['panel'] + [''] * 9


buttons = []
button_map = {}  
for i in range(3):
    row = []
    for j in range(3):
        digit = i * 3 + j + 1
        button = Button(root, text=str(digit), width=10, height=5)
        button.grid(row=i, column=j)
        button_map[digit] = button
        row.append(button)
    buttons.append(row)


def win():

    for i in range(0, 9, 3):
        if panels[i+1] == panels[i+2] == panels[i+3] != '':
            return panels[i+1]
    
    for i in range(1, 4):
        if panels[i] == panels[i+3] == panels[i+6] != '':
            return panels[i]
   
    if panels[1] == panels[5] == panels[9] != '':
        return panels[1]
    if panels[3] == panels[5] == panels[7] != '':
        return panels[3]
    return ''


def checker(digit):
    global count, mark, digits
    
    mark = 'X' if count % 2 == 0 else 'O'
    
    if digit == 1 and 1 in digits:
        digits.remove(1)
        panels[1] = mark
        button_map[1].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 2 and 2 in digits:
        digits.remove(2)
        panels[2] = mark
        button_map[2].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 3 and 3 in digits:
        digits.remove(3)
        panels[3] = mark
        button_map[3].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 4 and 4 in digits:
        digits.remove(4)
        panels[4] = mark
        button_map[4].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 5 and 5 in digits:
        digits.remove(5)
        panels[5] = mark
        button_map[5].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 6 and 6 in digits:
        digits.remove(6)
        panels[6] = mark
        button_map[6].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 7 and 7 in digits:
        digits.remove(7)
        panels[7] = mark
        button_map[7].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 8 and 8 in digits:
        digits.remove(8)
        panels[8] = mark
        button_map[8].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    if digit == 9 and 9 in digits:
        digits.remove(9)
        buttons[2][2].config(text=mark, state='disabled')
        count += 1
        if win() == mark:
            showinfo("Game Over", f"Player{'1' if mark == 'X' else '2'} wins")
            root.destroy()
    
   
    if count > 8 and win() == '':
        showinfo("Game Over", "Match Tied")
        root.destroy()


for digit in range(1, 10):
    button_map[digit].config(command=lambda d=digit: checker(d))


root.mainloop()

```
